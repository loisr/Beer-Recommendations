```python
# move user_names to the front of the dataframe
mid = um['user_name']
um.drop(labels=['user_name'], axis=1,inplace = True)
um.insert(0, 'user_name', mid)
um.head()
```

```python
# join the two different dataframes together
result2 = pd.concat([um1, um], axis=0, sort=False)
```

```python
result2.head()
```

```python
# some of the user names have extra things at the end from the web address
# &order=dateD&view=R
```

```python
# make a new df to make changes to, so if there are mistakes I can go back to the orginal without reruning everything
test = result2
```

```python
# get rid of the extra bits in the user names
test['user_name'].replace(regex=True,inplace=True,to_replace=r'\&(?m)(?<=\b&).*$',value=r'')
```

```python
# lets take a look and make sure it worked
test.head()
```

```python
# fill the NaN values with 0
test = test.fillna(0)
```

```python
# make sure it worked
test.head()
```

```python
# there were some repeats in the user names, lets group them together and sum the results
test = test.groupby('user_name').sum()
```

```python
# this went from 2330 rows to 1146
test.shape
```

```python
# lets see if there are any rows that don't have any reviews
for x in test.sum(axis=1):
    if x == 0:
        print(x)
```

```python
# let's get rid of all the rows without any reviews
test = test[test.sum(axis=1)>0]
```

# Reshape Data for Surprise!

Surprise doesn't want to be fed in data from a sparce matrix, they want data in just a few columns (User ID, Item, Rating, Timestamp(optional)). 

```python
# make a copy of the data
utility_matrix_copy = test.copy()

#get rid of the user names
utility_matrix_copy_features = utility_matrix_copy.reset_index(drop=True)

utility_matrix_copy_features.head()
```

```python
utility_matrix_copy['UID'] = list(range(1, len(utility_matrix_copy)+1))
utility_matrix_copy = utility_matrix_copy.reset_index(drop=True)
utility_matrix_copy.head()
```

```python
#rename index column UID - for User ID
utility_matrix_copy_features.index.name = 'UID'
utility_matrix_copy_features.head()
```

```python
#from initial utility matrix, pivot and drop 0 ratings

long_matrix = pd.melt(utility_matrix_copy, 
        id_vars='UID', 
        value_vars=utility_matrix_copy_features,
        var_name='beer_name',
        value_name='rating')

long_matrix = long_matrix[(long_matrix != 0).all(1)]
long_matrix.head()
```

```python
# reset index
long_matrix.reset_index(drop = True)
```

```python
# save data to csv
long_matrix.to_csv("./rating.csv", sep=',',index=False)
```
