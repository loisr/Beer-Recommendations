# Scrape Data about the beers that were reviewed
Name, brewery, style, abv

```python
counter = 0
#ba = pd.DataFrame()

all_beers = []
for user in user_profiles[8002::]:
    counter += 1
    
    #scraping
    url_page = requests.get(user)
    bf_scrape = BeautifulSoup(url_page.content, 'html.parser') 

    beer_dict = {}
    nb_list = []
    type_list = []
    abv = []
    keys = ['beer_name', 'brewery', 'style', 'abv']
    
    beer_name_brew = bf_scrape.select('a[href*=/beer/profile]')
    beer_type = bf_scrape.select('a[href*=/beer/styles/]')[2:]
    beer_abv = bf_scrape.select('td[valign]')[4::7]

    for x in beer_name_brew:
        nb_list.append(x.text)
    for x in beer_type:
        type_list.append(x.text)
    for x in beer_abv:
        abv.append(x.text)

    name_text = nb_list[0::2]
    brew_text = nb_list[1::2]
    type_text = type_list
    abv_text = abv
    
    beer_dict = dict(zip(keys, [name_text, brew_text, type_text, abv_text]))
    ba = pd.concat([ba, pd.DataFrame(beer_dict)], axis=0)
    
    ba = ba.reset_index().drop(["index"], axis=1)
    
    if counter % 100 == 0:         
        print(counter)
        ba.to_json("./all_beers_ba9.json")     
```
