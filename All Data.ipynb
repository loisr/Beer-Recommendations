```python
#find the duplicate beers
beer_counts = {}
for beer in list(result1['name']):
    if not beer in beer_counts:
        beer_counts[beer] = 1
    else:
        beer_counts[beer] += 1

duplicated = []
for name, count in beer_counts.items():    
    if count > 1 or count <1:
        duplicated.append(name)
```

```python
duplicated
```

```python
# get rid of duplicate beers
beer_matrix = result1[~result1['name'].isin(duplicated)]
beer_matrix = beer_matrix.reset_index()
```

Its really interesting that 3 of the top 10 styles are IPAs, and 4 are Stouts, lets figure out the counts of more general styles of beer.

```python
all_styles = dict(Counter(type_counts))
```

```python
all_styles
```

```python
count = 0 
for key, value in all_styles.items():
    if 'IPA' in key:
        count = count + value
print(count)
```

```python
# these are styles that seem to repeat in the data
types = ['IPA', 'Stout', 'Saison', 'Lambic', 'Barleywine', 'Porter', 'Brown Ale', 'Lager', 'Blonde Ale', 
         'Gose', 'Red Ale', 'Pilsener', 'Wheat']
```

```python
overall_types = {}
for n in types:
    count = 0 
    for key, value in all_styles.items():
        if n in key:
            count = count + value
    overall_types[n] = count
```

```python
overall_types
```
